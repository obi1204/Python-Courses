```python
#Challenge 1: Handling Missing Values
#import pandas and numpy
import numpy as np
import pandas as pd

sfPermits = pd.read_csv('Building_Permits.csv')

```

```python
# set seed for reproduciablilty
np.random.seed(0)
sfPermits.sample(6)
```

```python
#Calculate total number of cells in dataframe
totalCells = np.product(sfPermits.shape)

#count number of missing values per column
missingCount = sfPermits.isnull().sum()

#Calculate total number of missing values
totalMissing = missingCount.sum()

#Calculate percentage of missing values
round(((totalMissing/totalCells)*100), 2)
```

```python
#removing rows with missing values
sfPermits.dropna()
```

```python
#drop columns with empty values
sfPermitsCleanCols = sfPermits.dropna(axis=1)
#how much data is left
sfPermitsCleanCols
```
