```python
#import pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

#import amazon historical stock dataset
amazon = pd.read_csv("Amazon.csv")
```

```python
amazon
```

```python
#data in descending order
desc = amazon.sort_values("High")
desc
```

```python
#frequency of high prices
desc["High"].plot.hist();
```

```python
#converting date column to datetime
amazon["Date"] = pd.to_datetime(amazon['Date'], infer_datetime_format=True)
```

```python
amazon.plot.scatter(x = "High", y = "Date");
```

```python
#statistical info of High column
amazon["High"].describe()
```

```python
#importing gamespot data
gstop = pd.read_csv("GME_stock.csv")
gstop
```

```python
gdesc = gstop.sort_values("date")
gdesc
```

```python
gstop['date'] = pd.to_datetime(gstop['date'], infer_datetime_format=True)
```

```python
#comparing Gamestop and Amazon High prices
y1, y2= amazon["High"], gstop["high_price"]
x1, x2 = amazon['Date'], gstop['date']

fig, ax = plt.subplots()
ax.plot(x1,y1, x2,y2)

plt.show;
```
