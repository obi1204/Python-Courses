```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
uk = pd.read_csv("smoking.csv")
```

```python
uk
```

```python
#creating table of smokers
smoke = uk[uk.smoke == 'Yes']
smoke
```

```python
#more women smoke than men in this survey
plt.figure(figsize=(10,8))
plt.title('Gender Count')
sns.countplot(y='gender', data=smoke)
plt.show()
```

```python
#Twenty-eight year olds smoke the most
plt.figure(figsize=(10,15))
plt.title('Age Count')
sns.countplot(y='age', data=smoke)
plt.show()
```

```python
gm = pd.crosstab(smoke['gender'], smoke['marital_status'], )

gm.plot(kind='barh',stacked=True,figsize=(10,8))
plt.title('Gender by Marital Status')
plt.show()
```

```python
#those with no degree tend to smoke more,
gm = pd.crosstab(smoke['gross_income'], smoke['highest_qualification'])

gm.plot(kind='barh',stacked=True,figsize=(10,8))
plt.title('Gender by Marital Status')
plt.show()
```

```python
gm = pd.crosstab(smoke['nationality'], smoke['ethnicity'])

gm.plot(kind='barh',stacked=True,figsize=(10,8))
plt.title('Nationality by Ethinicity')
plt.show()
```

```python
gm = pd.crosstab(smoke['region'], smoke['nationality'])

gm.plot(kind='barh',stacked=True,figsize=(10,8))
plt.title('Region by Nationality')
plt.show()
```
