```python
#importing crime data from Austin, Houston, Fort Worth and Denton
import pandas as pd

denton_crime = pd.read_csv("Denton Crime Data.csv")
fort_worth_crime = pd.read_csv("Fort Worth Crime 2022.csv")
austin_crime = pd.read_csv("Austin Crime 2022.csv")
houston_crime = pd.read_excel("Houston Crime 2022.xlsx")
```

```python
#adding city column to Austin table
austin_crime['City'] = 'AUSTIN'
austin_crime
```

```python
#removing unnecessary columns so tables will match
updated1_austin = austin_crime.drop(columns = ['Incident Number', 'Highest Offense Code', 'Family Violence', 'Occurred Date Time', 'Occurred Time', 'Report Date Time', 'Report Date', 'Report Time', 'Address', 'Council District', 'APD Sector', 'APD District', 'PRA', 'Census Tract', 'Clearance Status', 'Clearance Date', 'UCR Category', 'Category Description', 'X-coordinate', 'Y-coordinate', 'Latitude', 'Longitude', 'Location'])
updated1_austin
```

```python
updated1_houston = houston_crime.drop(columns = ['Incident', 'RMSOccurrenceHour', 'NIBRSClass', 'OffenseCount', 'Beat', 'StreetNo', 'StreetName', 'StreetType','Suffix', 'MapLongitude', 'MapLatitude'])
updated1_houston
```

```python
denton_crime
```

```python
updated1_denton = denton_crime.drop(columns = ["id", 'agency', 'public_address'])
updated1_denton
```

```python
updated1_denton['City'] = 'DENTON'
updated1_denton
```

```python
fort_worth_crime
```

```python
updated1_fortworth = fort_worth_crime.drop(columns = ['Case and Offense', 'Case Number', 'Offense', 'Block Address', 'State', 'Beat', 'Council District', 'Attempt Complete', 'Location Type'])
updated1_fortworth
```

```python
updated1_austin
```

```python
updated1_austin.drop(columns = ["Zip Code"])
```

```python
drop_hou = updated1_houston.drop(columns = ["ZIPCode"])
```

```python
new_columns = ['Occurred Date', 'Highest Offense Description', 'Location Type', 'City']
updated1_austin = updated1_austin[new_columns]
updated1_austin
```

```python
#renaming colums
final1 = updated1_austin.rename(columns = {'Occurred Date': 'Date_of_Occurrence', 'Highest Offense Description':'Offense_Description', 'Location Type': 'Location_Type', 'City':'City'})
final2 = drop_hou.rename(columns = {'RMSOccurrenceDate': 'Date_of_Occurrence', 'NIBRSDescription':'Offense_Description', 'Premise': 'Location_Type', 'City':'City'})
```

```python
final2
```

```python
final1
```

```python
#reseting indexes before merge
final1.reset_index(drop=True, inplace=True)
final2.reset_index(drop=True, inplace=True)
```

```python
#merging tables
merge_austin_houston = pd.concat([final2, final1])
```

```python
merge_austin_houston
```

```python
updated1_denton
```

```python
final3 = updated1_denton.rename(columns = {'date/time': 'Date_of_Occurrence', 'location_name': 'Location_Type', 'crime' : 'Offense_Description'})
final3
```

```python
new_columns2 = ['Date_of_Occurrence', 'Offense_Description', 'Location_Type', 'City']
final_3 = final3[new_columns2]
final_3
```

```python
updated1_fortworth
```

```python
final4 = updated1_fortworth.rename(columns = {'From Date': 'Date_of_Occurrence', 'Location Description': 'Location_Type', 'Nature Of Call' : 'Offense_Description'})
final4
```

```python
drop_fortworth = final4.drop(columns = {"Reported Date", "Description", "Division", "Location"})
drop_fortworth
```

```python
new_columns3 = ['Date_of_Occurrence', 'Offense_Description', 'Location_Type', 'City']
final_4 = drop_fortworth[new_columns3]
final_4
```

```python
#resetting index
final_3.reset_index(drop=True, inplace=True)
final4.reset_index(drop=True, inplace=True)
```

```python
#merging all tables
complete = pd.concat([merge_austin_houston, final_3, final_4])
complete
```

```python
#making contents of columns same format, all capitalized
complete['Offense_Description'] = complete['Offense_Description'].str.upper()
```

```python
complete['Location_Type'] = complete['Location_Type'].str.upper()
```

```python
complete = complete.rename(columns={'Date_of_Occurence': 'Date'})
complete
```

```python
#converting date column from datetime to date
complete['Date_of_Occurrence'] = pd.to_datetime(complete['Date_of_Occurrence']).dt.date
```

```python
complete
```

```python
complete = complete.rename(columns={'Date_of_Occurrence': 'Date'})
```
