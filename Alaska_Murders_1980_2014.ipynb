```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
murders = pd.read_csv('database.csv')
```

```python
murders.info()
```

```python
murders
```

```python
alaska = murders[murders.State == 'Alaska']
alaska
```

```python
alaska.rename(columns = {'Victim Sex': 'Victim_Sex'})
alaska.info()
```

```python
alaska.drop(columns = ['Record ID', 'Agency Code','Record Source', 'Incident'])
```

```python
plt.figure(figsize=(10,8))
plt.title('Crime Solved?')
sns.countplot(y='Crime Solved', data=alaska)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title('Crime Solved?')
sns.countplot(y='Victim Sex', data=alaska)
plt.show()
```

```python
gm = pd.crosstab(alaska['Victim Sex'], alaska['Crime Type'], )

gm.plot(kind='barh',stacked=True,figsize=(10,8))
plt.title('Crime Type by Sex')
plt.show()
```

```python
plt.figure(figsize= (10,8))
plt.title("Murder Rate by Year")
sns.countplot(y='Year', data=alaska)
plt.show()
```

```python
Femicide= alaska[alaska['Victim Sex'] == 'Female']
```

```python
plt.figure(figsize=(10,8))
plt.title('Crime Solved?')
sns.countplot(y='Crime Solved', data=Femicide)
plt.show()
```

```python
sns.lineplot(x='Year', y='Crime Type', data=Femicide)
```

```python
plt.figure(figsize=(10,20))
plt.title("Age")
sns.countplot(y='Victim Age', data=Femicide)
plt.show()
```

```python
Femicide["Victim Age"].mode()
```

```python
gm = pd.crosstab(Femicide['Victim Age'], Femicide['Crime Type'])

gm.plot(kind='barh',stacked=True,figsize=(10,20))
plt.title('Age by Crime Type')
plt.show()
```
